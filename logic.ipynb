```python
# Wright a program to Calculating taxes according to annual income.

def calculate_tax(income):
    if income <= 500000:
        return 0
    elif income <= 1000000:
        return (income - 500000) * 0.10
    elif income <= 2000000:
        return (500000 * 0.10) + (income - 1000000) * 0.20
    else:
        return (500000 * 0.10) + (1000000 * 0.20) + (income - 2000000) * 0.30

salary = int(input("Write down your annual income (in rupees):"))
tax = calculate_tax(salary)
print("Tax to be Payed :Rs", tax)
```

```python
#wright a program You can create, view, and delete your own to-do lists.
tasks = []

def show_tasks():
    print("todays work\n")
    for i, task in enumerate(tasks, 1):
        print(f"{i}. {task}")

def add_task(task):
    tasks.append(task)

def remove_task(index):
    if 0 < index <= len(tasks):
        tasks.pop(index - 1)

while True:
    print("\n1Add new task\n2. Show task\n3. Finish task\n4. Close")
    choice = input("Choose!")

    if choice == "1":
        task = input("enter work:")
        add_task(task)
    elif choice == "2":
        show_tasks()
    elif choice == "3":
        show_tasks()
        index = int(input("which number to remove ?"))
        remove_task(index)
    elif choice == "4":
        break
    else:
        print("wrong choice")
```

```python
# Wright a program to find your password strong or weak?

def is_strong(Password):
    if len(password) < 8:
        return False
    has_digit = any(char.isdigit() for char in password)
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_special = any(char in "!@#$%^&*()-_=+[]{}" for char in password)

    return has_digit and has_upper and has_lower and has_special

pwd = input("enter your password:")

if is_strong(pwd):
    print("Strong psaaword:")
else:
    print("wrong password_ pleas try again:")
```
